## 1. Write a function that returns the greater of two numbers

```python
def greater(a,b):
    
    if a>b:
        return a
    else:
        return b
    


```

## 2. Now write a function that returns the largest element on a list

```python
def greatest(arr):
    
    max=0
    
    for e in arr:
        if max<e:
            max=e
    
    return max

```

```python
greatest([1,3,35,5656,4543,345,45,4])
```

```python
# This will test your function 
test_greatest(greatest)
```

## 3. Write a function that sums all the elements on a list

```python
def sum_all(arr):
    suma=0   
    for e in arr:
        suma+=e
        
    return suma
        
```

## 5. Now combine those two ideas and write a function that receives a list and either "+" or "*" and outputs acordingly

```python
def oper_all(arr, oper):
    
    sumar=0
    multip=1
    
    for e in arr:
    
        if oper=='+':
            sumar+=e
    
        if oper=='*':
            multip*=e
        
    if oper=='+':
        return sumar
    
    if oper=='*':
        return multip
        
```

```python
oper_all([5,6,3,2],'*')
```

```python
# This will test your function 
test_operations(oper_all)
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    
    if n==1 or n==0:
        return 1
    if n>=1:
        return n*factorial(n-1)
        
```

```python
factorial(6)
```

```python
# This will test your function 
test_factorial(factorial)
```

## 7. Write a function that takes a list and returns a list of the unique values.

`NOTE: You cannot use set. ðŸ¤”`

```python
def unique(arr):
    
    lst=[]
    
    for i in arr:
        if i not in lst:
            lst.append(i)
    return lst
    
    
```

```python
unique([1,3,4,3,2,2,3,5,8,9,3,5,2])
```

```python
# This will test your function 
test_unique(unique)
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... ðŸ§`

```python
def mode_counter(arr):
    
    freq={}
    
    for n in arr:
        if n in freq:
            freq[n]+=1
        else:
            freq[n]=1
            
    moda=0
    numero=0
    
    for k,v in freq.items():
        if v>moda:
            moda=v
            numero=k
    
    return numero
        
            
    
            
        
            
            
```

```python
mode_counter([1,3,4,4,4,4,4,4,4,4,3,2,2,3,5,8,9,3,5,2])
```

```python
# This will test your function 
test_mode(mode_counter)
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
    
    alph= "abcdefghijklmnopqrstuvwxyz"
    
    for char in alph:
        if char not in string.lower():
            
            return False
    
    
    return True
```

```python
# This will test your function 
test_pangram(pangram)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.
`Valid special characters: # @ ! $ % & ( ) ^ * [ ] { }`

```python
def check_pass(string):

      
    specialchar =['#','@','!','$','%','&','( )','^','*','[ ]','{ }'] 
    val = True
      
    if len(string) < 8: 
        
        print('length should be at least 8') 
        
        val = False
        

        
          
    if not any(char.isdigit() for char in string): 
        
        print('Password should have at least one numeral') 
        
        val = False
        
     
          
    if not any(char.isupper() for char in string): 
        
        print('Password should have at least one uppercase letter') 
        
        val = False

          
    if not any(char.islower() for char in string): 
        
        print('Password should have at least one lowercase letter') 
        
        val = False
        
   
          
    if not any(char in specialchar for char in string): 
        
        print('Password should have at least one of the symbols special character') 
        
        val = False
```

```python
# This will test your function 
test_pass(check_pass)
```
