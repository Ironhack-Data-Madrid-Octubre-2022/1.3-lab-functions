## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    
    factorial = 1
    
    for i in range(1,n + 1): 
        factorial *= i
    return factorial
        
```

## 7. Write a function that takes a list and returns a list of the unique values.

`NOTE: You cannot use set. ü§î`

```python
def unique(arr):
    
    unique = []
    
    for i in arr:
        if i not in unique:
            unique.append(i)
    
    return unique
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... üßê`

```python
def mode_counter(arr):
    
    count = {}  #creamos diccionario
    
    for i in arr:
        if i not in count:     #si i no aparece previamente en la lista pon 1
            count[i] = 1
        
        else:                  #si i aparece previamente en la lista suma 1
            count[i] += 1
            
    
    # y es la variable con la que ordenamos el diccionario por la max variable
    # reverse = True, para cambiar el orden y ponerlo en descendente
    # .keys(), para imprimir solo las keys
    
    y={k: v for k, v in sorted(count.items(), key=lambda item: item[1], reverse=True)}.keys()  
    
    # luego trasformo a lista
    y=list(y)
    
    # y por ultimo pido que me devuelva solo el primer valor de la lista
    return y[0]
           
```

```python
# This will test your function 
test_mode(mode_counter)
```
